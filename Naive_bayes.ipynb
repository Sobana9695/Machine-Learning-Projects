```python
import re
import nltk 
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import pandas as pd
data=pd.read_csv(r"C:\Users\SSS\Downloads\bbc-text.csv")
data
```

```python
fv=data[["text"]]
cl=data[["category"]]
```

```python
fv.head(2)
```

```python
cl.head()
```

```python
stp=stopwords.words("english")    # default language-urdu
stp.remove("not")
```

```python
"not" in stp
```

```python
lt=[]
for word in data["text"]:
    l=[]
    ls=word.lower()
    ln=re.sub("[^a-zA-Z]"," ",ls)
    for s in word_tokenize(ln):
        if s in stp or len(s)<=2:
            pass
        else:
            ps=PorterStemmer().stem(s)
            l.append(ps) 
    lt.append(" ".join(l))
```

```python
fv=pd.DataFrame({"cleaned_text":lt})
fv
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(fv,cl,train_size=0.8)
```

```python
x_train.shape
```

```python
cv=CountVectorizer()
```

```python
dtm=cv.fit_transform(x_train["cleaned_text"])
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB,GaussianNB,CategoricalNB
```

```python
mnb=MultinomialNB()
```

```python
model=mnb.fit(dtm,y_train)
```

```python
x_test
```

```python
x_test1=cv.transform(x_test["cleaned_text"])
```

```python
predict=model.predict(x_test1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predict)
```
